```python
import random
def randomPerm(n):
    v=[]
    for i in range(n):
        v.append(i+1)
    for i in range(len(v)-1):
        j = random.randint(i, len(v)-1)
        aux = v[i]
        v[i] = v[j]
        v[j] = aux
    return v

def bubbletSteps(a):
    done = False
    steps = 0
    while not done:
        done = True
        for i in range(len(a)-1):
            if a[i] > a[i+1]:
                done = False
                aux = a[i]
                a[i] = a[i+1]
                a[i+1] = aux
                steps += 3
            steps += 1
        steps += 1
    return steps


n = 100
runs = 10000
arr = []
sum = 0
min = n**3
max = 0
for i in range(runs):
    t = bubbletSteps(randomPerm(n))
    arr.append(t)
    if t < min :
       min = t
    if t > max :
       max = t
    sum = sum + t
print 'Experimenal best time, ' + str(min)
print 'Experimenal worst time,' + str(max)
print 'Experimenal average time,' + str(sum/runs)
```

```python
import numpy as np
numero = np.zeros((10000))
for i in range(10000):
    numero[i] = i
```
