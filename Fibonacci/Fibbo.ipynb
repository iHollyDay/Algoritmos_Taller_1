```python
import time
timeArreglo = []
numero = []
def fibbo(n):
    start_time = time.time()
    a=0
    b=1
    f=0
    for i in range(int(n)-1):
        a=b
        b=f
        f=a+b
    timef = time.time() - start_time
    print(str(timef) +"\t"+str(n))
    timeArreglo.append(timef)
    numero.append(n)

for i in range(0,500000, 1000):
    fibbo(i)
```

```python
import matplotlib.pyplot as plt
plt.plot(numero, timeArreglo, 'ro')
plt.ylabel('time')
plt.xlabel('n')
plt.show()
```
